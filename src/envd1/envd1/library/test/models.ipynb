```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import PolynomialFeatures
```

```python
def generate_data():
    X = np.arange(-30, 30, 1)
    y = (9*X**3 - 5*X**2 + np.random.randn(60)*10000/.5)
    return X, y
trX, trY = generate_data()
trX = (trX- min(trX))/(max(trX)-min(trX))
trY = (trY- min(trY))/(max(trY)-min(trY))
#plot the data
fig, ax = plt.subplots()
ax.scatter(trX, trY)
plt.show()
```

```python
n = 3
trX_expanded = np.expand_dims(trX, axis=1)

poly = PolynomialFeatures(n)
# returns: [1, x, x^2, x^3]
trX_expanded = poly.fit_transform(trX_expanded)
```

## linear

```python
inp = Input((1)) 
out = Dense(1)(inp)
model_linear = Model(inputs=inp, outputs=out)
model_linear.compile(
optimizer=Adam(learning_rate=1e-3), 
loss="mean_squared_error")
```

```python
model_linear.fit(trX, trY, epochs=500)
```

```python
ax.scatter(trX, trY)
ax.plot(trX, model_linear.predict(trX), color="red")
plt.show()
```

## poly

```python
inp = Input((n+1)) 
#since one of the features is 1, we need an extra input
x = Dense(10, activation='relu')(inp)
out = Dense(1)(x)
model = Model(inputs=inp, outputs=out)
model.compile(optimizer=Adam(learning_rate=1e-3), loss="mean_squared_error")
```

```python
model.fit(trX_expanded, trY, epochs=700, batch_size=10);
```

```python
ax.scatter(trX, trY)
ax.plot(trX, model.predict(trX_expanded), color="blue")
plt.show()
```
